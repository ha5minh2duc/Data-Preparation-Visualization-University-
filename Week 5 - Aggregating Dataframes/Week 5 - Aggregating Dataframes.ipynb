### Ex1. Load dataset nyc_weather_2018 and weather_stations, then calculate the average elevation of stations that calculat‘SNOW’ and contains ‘NJ US’ in its name.

```python
weather = pd.read_csv('nyc_weather_2018.csv')
stations = pd.read_csv('weather_stations.csv')
```

```python
weather.head()
```

```python
weather.shape
```

```python
stations.head()
```

```python
stations.shape
```

```python
stations.rename(columns={"id":"station"}, inplace=True)
stations.head()
```

```python
merged_data = stations.merge(weather, on="station", how='inner').drop_duplicates(["station","datatype"]).reset_index(drop=True)
merged_data.head()
```

```python
merged_data.head()
```

```python
merged_data.shape
```

```python
# Calculate the average elevation of stations that calculate ‘SNOW’ and contains ‘NJ US’ in its name.
snow = merged_data[(merged_data.datatype == "SNOW") & (merged_data.name.str.contains("NJ US"))]
snow.head()
```

```python
snow.elevation.mean()
```

### Ex2. Load the fb dataset, then: 1) Calculate the percentage change between that and and last day in volumn (pct_change method). 2) Sort the dataframe by percentage change then get five days that have the biggest change. 

```python
fb = pd.read_csv('fb.csv')
fb.head()
```

```python
fb.shape
```

```python
# 1) Calculate the percentage change between that and and last day in volumn (pct_change method)
pct_change = fb.assign(pct_change = lambda x: x.volume.pct_change())
pct_change.head()
```

```python
# 2) Sort the dataframe by percentage change then get five days that have the biggest change
pct_change.sort_values(by='pct_change', ascending=False).head()
```

### Ex3. Create bins for each full number of earthquake magnitude (for instance, the first bin is (0, 1], the second is (1, 2], and so on) with the ml magnitude type and count how many are in each bin.

```python
earthquakes = pd.read_csv('earthquakes.csv')
earthquakes.head()
```

```python
earthquakes.shape
```

```python
earthquakes_ml = earthquakes[earthquakes.magType=='ml']
earthquakes_bins = pd.cut(earthquakes_ml.mag, [0, 1, 2, 3, 4, 5])
earthquakes_bins.value_counts()
```

### Ex4. Load the dataset weather_by_station then: 1) Calculate the mean value each station recorded snow per month. 2) Count the number of times each station recorded snow per month.

```python
weather = pd.read_csv('weather_by_station.csv')
weather.head()
```

```python
weather.shape
```

```python
weather = weather.assign(month=pd.DatetimeIndex(weather.date).month)
weather.head()
```

```python
# 1) Calculate the mean value each station recorded snow per month
pd.crosstab(index=weather.station, columns=weather.month, colnames=['Month'], values=weather.value, aggfunc=np.mean)
```

```python
# 2) Count the number of times each station recorded snow per month
pd.crosstab(index=weather.station, columns=weather.month, colnames=['Month'])
```
