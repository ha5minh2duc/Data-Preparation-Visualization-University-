## Ex1.

##### We want to look at data for the Facebook, Apple, Amazon, Netflix, and Google (FAANG) stocks. Combine them into a single file and store the dataframe of the FAANG data as faang for the rest of the exercises:

```python
# a) Read in the aapl.csv, amzn.csv, fb.csv, goog.csv, and nflx.csv files.
```

```python
aapl = pd.read_csv("aapl.csv")
amzn = pd.read_csv("amzn.csv")
fb = pd.read_csv("fb.csv")
goog = pd.read_csv("goog.csv")
nflx = pd.read_csv("nflx.csv")
```

```python
aapl.head()
```

```python
amzn.head()
```

```python
fb.head()
```

```python
goog.head()
```

```python
nflx.head()
```

```python
# b) Add a column to each dataframe, called ticker, indicating the ticker symbol it is for (Apple's is AAPL, for example); 
# this is how you look up a stock. In this case, the filenames happen to be the ticker symbols.
```

```python
amzn = amzn.assign(ticker = 'AMZ')
fb = fb.assign(ticker = 'FB')
goog = goog.assign(ticker = 'GOOG')
nflx = nflx.assign(ticker = 'NFLX')
aapl = aapl.assign(ticker = 'AAPL')
```

```python
goog.head()
```

```python
# c) Append them together into a single dataframe. (concat)
```

```python
faang = pd.concat([amzn, fb, goog, nflx, aapl])
faang
```

```python
# d) Save the result in a CSV file called faang.csv. (to_csv)
```

```python
faang.to_csv('faang.csv', index=True)
```

```python
data = pd.read_csv("faang.csv", index_col=[0])
data.head()
```

```python
# e) With faang, use type conversion to cast the values of the date column into datetimes and the volume column into integers. 
# Then, sort by date and ticker.
```

```python
data = data.assign(date = lambda x: pd.to_datetime(x.date), 
                   volume = lambda x: x.volume.astype(int))
data
```

```python
data = data.sort_values(by=['date','ticker'], ascending=[True,True])
data.head()
```

```python
# f) Find the seven rows in faang with the lowest value for volume
```

```python
data = data.sort_values(by=['volume'], ascending=True)
data.head()
```

##  Ex2.

##### The European Centre for Disease Prevention and Control (ECDC) provides an open dataset on COVID-19 cases called daily number of new reported cases of COVID-19 by country worldwide. This dataset is updated daily, but we will use a snapshot that contains data from January 1, 2020 through September 18, 2020. Clean and pivot the data so that it is in wide format: 

```python
# a) Read in the covid19_cases.csv file
```

```python
df = pd.read_csv('covid19_cases.csv')
df.head()
```

```python
# b) Create a date column using the data in the dateRepcolumn and the pd.to_datetime() function. 
```

```python
df = df.assign(date = lambda x: pd.to_datetime(x.dateRep))
df.head()
```

```python
# c) Set the date column as the index and sort the index.
```

```python
df.set_index('date', inplace=True)
df.head()
```

```python
# d) Replace all occurrences of United_States_of_America and United_Kingdom with USA and UK, respectively. 
# Hint: the replace() method can be run on the dataframe as a whole.
```

```python
df.replace({'United_States_of_America':"US", "United_Kingdom":"UK"}, inplace=True)
df[df.countriesAndTerritories == 'UK'].head()
```

```python
# e) Using the countriesAndTerritoriescolumn, filter the cleaned COVID-19 cases data 
# down to Argentina, Brazil, China, Colombia, India, Italy, Mexico, Peru, Russia, Spain, Turkey, the UK, and the USA
```

```python
cleaned = df.loc[df.countriesAndTerritories.isin(['Argentina', 'Brazil', 'China', 'Colombia', 'India', 'Italy', 'Mexico', 'Peru', 'Russia', 'Spain', 'Turkey', 'UK', 'USA'])]
cleaned.head()
```

```python
# f) Pivot the data so that the index contains the dates, the columns contain the country names, and the values are the case counts (the cases column). 
# Be sure to fill in NaNvalues with 0.
```

```python
pivot_data = cleaned.pivot(columns='countriesAndTerritories', values='cases').fillna(0)
pivot_data.head()
```
